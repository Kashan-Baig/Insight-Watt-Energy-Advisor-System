```python
import pandas as pd
import numpy as np
import calendar




def extract_behavior_features(df):
    df = df.copy()

    # ---------- BASIC TIME FEATURES ----------
    df["hour"] = df.index.hour
    df["weekday"] = df.index.weekday
    df["is_weekend"] = df["weekday"] >= 5
    df["date"] = df.index.date
    df["month"] = df.index.month

    features = {}

    # ---------- PEAK HOURS ----------
    hourly_avg = df.groupby('hour')['use_at_kw'].mean()
    threshold = 0.8 * hourly_avg.max()
    peak_hours = hourly_avg[hourly_avg >= threshold].index.tolist()

    features["peak_hours"] = peak_hours
    features["peak_hours_consistent"] = True

    # ---------- PEAK MONTHS ----------
    monthly_kwh = df.groupby("month")["use_at_kw"].sum()
    top_months = monthly_kwh.nlargest(3)

    features["peak_months"] = [calendar.month_name[m] for m in top_months.index]
    features["monthly_kwh"] = {
        calendar.month_name[m]: round(v, 2) for m, v in top_months.items()
    }

    # ---------- SPIKE SUMMARY ----------
    spike_threshold = df["use_at_kw"].mean() + 2 * df["use_at_kw"].std()
    df["spike"] = df["use_at_kw"] > spike_threshold

    spike_count = int(df["spike"].sum())
    features["spike_count"] = spike_count

    if spike_count > 0:
        features["avg_spike_kwh"] = round(df[df["spike"]]["use_at_kw"].mean(), 2)

        spike_hours = df[df["spike"]].groupby("hour")["spike"].sum()
        spike_peak_hours = spike_hours[
            spike_hours >= 0.5 * spike_hours.max()
        ].index.tolist()

        features["spike_peak_window"] = spike_peak_hours

        weekend_spikes = (df[df["spike"]]["weekday"] >= 5).sum()
        features["weekend_spike_percent"] = round(
            weekend_spikes / spike_count * 100, 1
        )
    else:
        features["avg_spike_kwh"] = 0
        features["spike_peak_window"] = "none"
        features["weekend_spike_percent"] = 0

    # ---------- WEEKEND vs WEEKDAY ----------
    daily = df.groupby(["date", "is_weekend"])["use_at_kw"].sum().reset_index()

    avg_weekday = daily[daily["is_weekend"] == False]["use_at_kw"].mean()
    avg_weekend = daily[daily["is_weekend"] == True]["use_at_kw"].mean()

    weekend_delta = (
        (avg_weekend - avg_weekday) / avg_weekday * 100
        if avg_weekday > 0 else 0
    )

    if weekend_delta > 10:
        behavior = "weekend-heavy"
    elif weekend_delta < -10:
        behavior = "weekday-heavy"
    else:
        behavior = "balanced"

    features["weekend_behavior"] = behavior
    features["weekend_increase_percent"] = round(weekend_delta, 2)

    return features
```

```python
def extract_weather_features(df):
    features = {}

    # ---------- AVERAGES ----------
    features["avg_temp_c"] = round(df["Temperature"].mean(), 1)
    features["avg_humidity_percent"] = round(df["Humidity"].mean(), 1)
    features["avg_windspeed_kmph"] = round(df["Wind Speed"].mean(), 1)

    # ---------- HEAT STRESS INDEX (Simple) ----------
    # Not full HI formula — lightweight & hackathon-safe
    heat_stress = df["Temperature"] + 0.1 * df["Humidity"]
    features["avg_heat_stress_index"] = round(heat_stress.mean(), 1)

    # ---------- TEMP–ENERGY CORRELATION ----------
    corr = df["Temperature"].corr(df["use_at_kw"])
    features["temp_kwh_correlation"] = round(corr, 2)

    # ---------- HUMIDITY LABEL ----------
    if features["avg_humidity_percent"] > 70:
        humidity_label = "high"
    elif features["avg_humidity_percent"] > 50:
        humidity_label = "moderate"
    else:
        humidity_label = "low"

    features["humidity_level"] = humidity_label

    # ---------- WIND COOLING EFFECT ----------
    if features["avg_windspeed_kmph"] > 10:
        wind_effect = "good-ventilation"
    elif features["avg_windspeed_kmph"] > 5:
        wind_effect = "moderate"
    else:
        wind_effect = "poor-ventilation"

    features["wind_cooling_effect"] = wind_effect

    # ---------- WEATHER DRIVER ----------
    if corr > 0.4:
        driver = "cooling-driven"
    elif corr < -0.4:
        driver = "heating-driven"
    else:
        driver = "weather-neutral"

    features["weather_driver"] = driver

    return features

weather_features = extract_weather_features(df)
weather_features
```

```python
def format_for_llm(behavior, weather, total_hours):
    payload = {
        "usage_behavior": {
            "peak_hours": behavior["peak_hours"],
            "peak_months": behavior["peak_months"],
            "weekend_behavior": behavior["weekend_behavior"],
            "weekend_increase_percent": float(behavior["weekend_increase_percent"])
        },

        "spike_profile": {
            "spike_rate_percent": round(
                behavior["spike_count"] / total_hours * 100, 1
            ),
            "avg_spike_kw": float(behavior["avg_spike_kwh"]),
            "spike_peak_hours": behavior["spike_peak_window"],
            "weekend_spike_percent": float(behavior["weekend_spike_percent"])
        },

        "weather_context": {
            "avg_temp_c": float(weather["avg_temp_c"]),
            "thermal_condition": (
                "hot" if weather["avg_temp_c"] >= 30 else
                "warm" if weather["avg_temp_c"] >= 25 else
                "mild"
            ),
            "humidity_level": weather["humidity_level"],
            "wind_cooling_effect": weather["wind_cooling_effect"],
            "heat_stress_index": float(weather["avg_heat_stress_index"]),
            "temp_kwh_correlation": float(weather["temp_kwh_correlation"]),
            "weather_driver": weather["weather_driver"]
        }
    }

    return payload
```

```python
def validate_and_standardize_columns(df):
    df = df.copy()

    column_aliases = {
        "datetime": ["datetime", "timestamp", "date_time", "time"],
        "use_at_kw": ["use_at_kw", "kwh", "energy", "power", "usage_kw"],
        "Temperature": ["Temperature", "temp", "temp_c", "temperature_c"],
        "Humidity": ["Humidity", "humidity", "rh"],
        "Wind Speed": ["Wind Speed", "wind_speed", "windspeed", "wind_kmph"]
    }

    rename_map = {}

    # ---------- HANDLE DATETIME ----------
    if not isinstance(df.index, pd.DatetimeIndex):
        for col in df.columns:
            if col in column_aliases["datetime"]:
                rename_map[col] = "datetime"
                break
    else:
        # already datetime indexed
        df.index.name = "datetime"

    # ---------- HANDLE OTHER COLUMNS ----------
    for standard_name, aliases in column_aliases.items():
        if standard_name == "datetime":
            continue
        for col in df.columns:
            if col in aliases:
                rename_map[col] = standard_name
                break

    df.rename(columns=rename_map, inplace=True)

    # ---------- SET DATETIME INDEX ----------
    if "datetime" in df.columns:
        df["datetime"] = pd.to_datetime(df["datetime"])
        df.set_index("datetime", inplace=True)

    # ---------- FINAL VALIDATION ----------
    required_cols = ["use_at_kw", "Temperature", "Humidity", "Wind Speed"]
    missing = [col for col in required_cols if col not in df.columns]

    if missing:
        raise ValueError(f"Missing required columns: {missing}")

    if not isinstance(df.index, pd.DatetimeIndex):
        raise ValueError("Datetime index could not be inferred or created.")

    return df
```

```python
import pandas as pd
import numpy as np
import calendar


def extract_llm_features(df):
    """
    Input:
        df : raw DataFrame with columns
             - datetime
             - use_at_kw
             - Temperature
             - Humidity
             - Wind Speed

    Output:
        dict : LLM-ready compact feature payload
    """

    # ---------- SET DATETIME INDEX ----------
    df = df.copy()
    if "datetime" in df.columns:
        df["datetime"] = pd.to_datetime(df["datetime"])
        df.set_index("datetime", inplace=True)
    else:
        # already indexed
        df.index = pd.to_datetime(df.index)

    total_hours = len(df)

    # ---------- BASIC TIME FEATURES ----------
    df["hour"] = df.index.hour
    df["weekday"] = df.index.weekday
    df["is_weekend"] = df["weekday"] >= 5
    df["date"] = df.index.date
    df["month"] = df.index.month

    payload = {}

    # =====================================================
    # 🔹 USAGE BEHAVIOR FEATURES
    # =====================================================

    # ---------- PEAK HOURS ----------
    hourly_avg = df.groupby("hour")["use_at_kw"].mean()
    threshold = 0.8 * hourly_avg.max()
    peak_hours = hourly_avg[hourly_avg >= threshold].index.tolist()

    # ---------- PEAK MONTHS ----------
    monthly_kwh = df.groupby("month")["use_at_kw"].sum()
    top_months = monthly_kwh.nlargest(3)

    # ---------- WEEKEND VS WEEKDAY ----------
    daily = df.groupby(["date", "is_weekend"])["use_at_kw"].sum().reset_index()
    avg_weekday = daily[daily["is_weekend"] == False]["use_at_kw"].mean()
    avg_weekend = daily[daily["is_weekend"] == True]["use_at_kw"].mean()

    weekend_delta = (
        (avg_weekend - avg_weekday) / avg_weekday * 100
        if avg_weekday > 0 else 0
    )

    if weekend_delta > 10:
        weekend_behavior = "weekend-heavy"
    elif weekend_delta < -10:
        weekend_behavior = "weekday-heavy"
    else:
        weekend_behavior = "balanced"

    payload["usage_behavior"] = {
        "peak_hours": peak_hours,
        "peak_months": [calendar.month_name[m] for m in top_months.index],
        "weekend_behavior": weekend_behavior,
        "weekend_increase_percent": round(float(weekend_delta), 2)
    }

    # =====================================================
    # 🔹 SPIKE PROFILE
    # =====================================================

    spike_threshold = df["use_at_kw"].mean() + 2 * df["use_at_kw"].std()
    df["spike"] = df["use_at_kw"] > spike_threshold
    spike_count = int(df["spike"].sum())

    if spike_count > 0:
        spike_hours = df[df["spike"]].groupby("hour")["spike"].sum()
        spike_peak_hours = spike_hours[
            spike_hours >= 0.5 * spike_hours.max()
        ].index.tolist()

        weekend_spikes = (df[df["spike"]]["weekday"] >= 5).sum()

        payload["spike_profile"] = {
            "spike_rate_percent": round(spike_count / total_hours * 100, 1),
            "avg_spike_kw": round(float(df[df["spike"]]["use_at_kw"].mean()), 2),
            "spike_peak_hours": spike_peak_hours,
            "weekend_spike_percent": round(
                weekend_spikes / spike_count * 100, 1
            )
        }
    else:
        payload["spike_profile"] = {
            "spike_rate_percent": 0,
            "avg_spike_kw": 0,
            "spike_peak_hours": [],
            "weekend_spike_percent": 0
        }

    # =====================================================
    # 🔹 WEATHER CONTEXT
    # =====================================================

    avg_temp = float(df["Temperature"].mean())
    avg_humidity = float(df["Humidity"].mean())
    avg_wind = float(df["Wind Speed"].mean())

    heat_stress = df["Temperature"] + 0.1 * df["Humidity"]
    temp_corr = df["Temperature"].corr(df["use_at_kw"])

    if avg_temp >= 30:
        thermal_condition = "hot"
    elif avg_temp >= 25:
        thermal_condition = "warm"
    else:
        thermal_condition = "mild"

    if avg_humidity > 70:
        humidity_level = "high"
    elif avg_humidity > 50:
        humidity_level = "moderate"
    else:
        humidity_level = "low"

    if avg_wind > 10:
        wind_effect = "good-ventilation"
    elif avg_wind > 5:
        wind_effect = "moderate"
    else:
        wind_effect = "poor-ventilation"

    if temp_corr > 0.4:
        weather_driver = "cooling-driven"
    elif temp_corr < -0.4:
        weather_driver = "heating-driven"
    else:
        weather_driver = "weather-neutral"

    payload["weather_context"] = {
        "avg_temp_c": round(avg_temp, 1),
        "thermal_condition": thermal_condition,
        "humidity_level": humidity_level,
        "wind_cooling_effect": wind_effect,
        "heat_stress_index": round(float(heat_stress.mean()), 1),
        "temp_kwh_correlation": round(float(temp_corr), 2),
        "weather_driver": weather_driver
    }

    return payload
```
