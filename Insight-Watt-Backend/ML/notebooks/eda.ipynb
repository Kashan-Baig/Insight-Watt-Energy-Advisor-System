```python
df['datetime'] = pd.to_datetime(df['datetime'])
df['month'] = df['datetime'].dt.month          # 1â€“12
df['day'] = df['datetime'].dt.day              # 1â€“31
df['is_weekend'] = df['datetime'].dt.dayofweek >= 5
df['is_weekend'] = df['is_weekend'].astype(int)
df['day_of_week'] = df['datetime'].dt.dayofweek
```

```python
import numpy as np
df['Hour_sin'] = np.sin(2 * np.pi * df['hour'] / 24)
df['Hour_cos'] = np.cos(2 * np.pi * df['hour'] / 24)
df['Month_sin'] = np.sin(2 * np.pi * df['month'] / 12)
df['Month_cos'] = np.cos(2 * np.pi * df['month'] / 12)
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler

features_to_scale = [
    'House_num',
    'hour',
    'month',
    'day',
    'day_of_week'
]

st = StandardScaler()
df[features_to_scale] = st.fit_transform(df[features_to_scale])
```

```python
df['usage_lag1'] = df['use_at_kw'].shift(1)
df['usage_lag2'] = df['use_at_kw'].shift(2)
df['usage_lag3'] = df['use_at_kw'].shift(3)
df['usage_lag4'] = df['use_at_kw'].shift(4)
df['usage_lag5'] = df['use_at_kw'].shift(5)
df['usage_lag6'] = df['use_at_kw'].shift(6)
```

```python
df['usage_roll3h'] = df['use_at_kw'].shift(1).rolling(window=24).mean()
```

```python
df = df.fillna(0)
```

```python
split_idx = int(len(df) * 0.8)
train_df = df.iloc[:split_idx]
test_df = df.iloc[split_idx:]

print(f"Train size: {len(train_df)} ({train_df['datetime'].min().date()} to {train_df['datetime'].max().date()})")
print(f"Test size: {len(test_df)} ({test_df['datetime'].min().date()} to {test_df['datetime'].max().date()})")


# Drop datetime from features but keep target
X_train = train_df.drop(columns=['use_at_kw', 'datetime','House_ID'])
y_train = train_df['use_at_kw']

X_test = test_df.drop(columns=['use_at_kw', 'datetime','House_ID'])
y_test = test_df['use_at_kw']
```

```python
import catboost
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score


model = catboost.CatBoostRegressor(
        iterations=300,
        learning_rate=0.05,
        depth=6,
        random_state=42,
        verbose=False
    )

model.fit(X_train, y_train)
y_pred = model.predict(X_test)
    
mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)

print(f"  MAE:  {mae:.4f}")
print(f"  RMSE: {rmse:.4f}")
print(f"  RÂ²:   {r2:.4f}")
```

```python
import numpy as np
import pandas as pd
import re

def forecast_power_from_datetime(
    input_datetime,
    house_id,
    df_history,
    model,
    scaler
):
    # -----------------------------
    # 0. Extract numeric House ID
    # -----------------------------
    # if isinstance(house_id, str):
    #     house_id_num = int(re.search(r'\d+', house_id).group())
    # else:
    #     house_id_num = int(house_id)

    # -----------------------------
    # 1. Parse datetime
    # -----------------------------
    dt = pd.to_datetime(input_datetime)

    hour = dt.hour
    month = dt.month
    day = dt.day
    day_of_week = dt.dayofweek
    is_weekend = int(day_of_week >= 5)

    # -----------------------------
    # 2. Cyclical features
    # -----------------------------
    Hour_sin = np.sin(2 * np.pi * hour / 24)
    Hour_cos = np.cos(2 * np.pi * hour / 24)

    Month_sin = np.sin(2 * np.pi * month / 12)
    Month_cos = np.cos(2 * np.pi * month / 12)

    # -----------------------------
    # 3. Historical data BEFORE datetime
    # -----------------------------
    hist = df_history[
        (df_history['House_ID'] == house_id) &
        (df_history['datetime'] < dt)
    ].sort_values('datetime')

    if hist.empty:
        raise ValueError(f"No historical data for House_ID {house_id}")

    # -----------------------------
    # 4. Lag features (safe padding)
    # -----------------------------
    recent_usage = hist['use_at_kw'].tail(6).values

    if len(recent_usage) < 6:
        pad_value = hist['use_at_kw'].mean()
        recent_usage = np.pad(
            recent_usage,
            (6 - len(recent_usage), 0),
            constant_values=pad_value
        )

    usage_lag1, usage_lag2, usage_lag3, usage_lag4, usage_lag5, usage_lag6 = recent_usage[::-1]
    usage_roll3h = np.mean([usage_lag1, usage_lag2, usage_lag3])

    house_id = int(re.search(r'\d+', house_id).group())
    # -----------------------------
    # 5. Feature vector (ORDER MATTERS)
    # -----------------------------
    X_input = np.array([[
        hour,
        month,
        day,
        day_of_week,
        house_id,
    ]])

    # -----------------------------
    # 6. Scale & Predict
    # -----------------------------
    to_scale = np.array([[hour, month, day, day_of_week, house_id]])
    scaled = scaler.transform(to_scale)[0]

    hour_s, month_s, day_s, dow_s, house_s = scaled

    X_input = np.array([[ 
        hour_s,
        month_s,
        day_s,
        dow_s,
        house_s,
        is_weekend,
        Hour_sin,
        Hour_cos,
        Month_sin,
        Month_cos,
        usage_lag1,
        usage_lag2,
        usage_lag3,
        usage_lag4,
        usage_lag5,
        usage_lag6,
        usage_roll3h
    ]])

    prediction = model.predict(X_input)

    return float(prediction[0])
```
